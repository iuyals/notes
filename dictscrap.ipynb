```
!wget http://d.noty50.top/static/files/7000words
import requests as httprequests
import json
import lxml.html


def rmHeadNumber(str):
    if (not str):
        return ''
    PointPos = 0
    while(str[PointPos] != '.'):
        if(PointPos >= len(str)-1):
            return str
        PointPos = PointPos+1

    return str[PointPos+1:]


def getCollinsTrees(keyword):
    url = "http://dict.youdao.com/w/eng/"+keyword
    page = lxml.html.parse(url)
    trees = page.xpath("//*[@id='collinsResult']//ul")
    return trees


def getDefineList(trees):
    strlist = []
    for t in trees:
        s = t.text_content()
        s = s.strip()
        s = s.replace("\t", "")
        s = s.replace("  ", "")
        s = s.replace("\n\n", "\n")
        s = s.replace("\n\n", "\n")
        s = s[0:12].replace("\n", " ")+s[12:]
        if len(s) > 0:
            s=rmHeadNumber(s)
            strlist.append(s)
    return strlist


def printStrlist(strlist):
    for s in strlist:
        print(s)


def lookup(keyword):
    trees = getCollinsTrees(keyword)
    strlist = getDefineList(trees)

    return strlist

dict={}
counter=1
with open('./7000words') as wtext:
    for li in wtext:
        word=li.replace('\n','')
        dict[word]=lookup(word)
        print(f'got word {word} complete {counter}')
        counter=counter+1

print('writing to file json')
with open('./dict.json','w') as fdictjson:
    fdictjson.write(json.dumps(dict))

print('complete')
```
